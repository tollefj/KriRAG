```python
import json
from tqdm.notebook import tqdm
import os

folder = "json"
files = os.listdir(folder)
files = [os.path.join(folder, f) for f in files]


file_contents = []
for file_path in tqdm(files):
    with open(file_path, "r") as f:
        data = json.load(f)
        paragraphs = data["paragraphs"].values()
        content = []
        # combine all "content" of each paragraphs
        for p in paragraphs:
            for c in p["content"]:
                content.append(c)
        # combine as newline-separated string
        content = "\n".join(content)
        file_id = file_path.split("/")[-1].split(".")[0] + ".txt"
        print(file_id)
        outfile = os.path.join("txt", file_id)
        os.makedirs(os.path.dirname(outfile), exist_ok=True)
        with open(outfile, "w") as out:
            out.write(content)
        
```
